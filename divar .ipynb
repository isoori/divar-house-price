```python
import pandas as pd
import numpy as np
from unidecode import unidecode
```

```python
df = pd.read_csv('housing-Raoofi.csv',low_memory=False)
df.head()
```

```python
df['Address'] = df['Address'].str.replace('،',' ')
```

```python
df['Price'] = df['Price'].str.replace('٬','')
df['Price'] = df['Price'].str.replace('تومان','')
```

```python
df.head()
```

```python
df['Price'] = df['Price'].apply(lambda x : unidecode(x))
df.head()
```

```python
df.drop(51,inplace=True)
```

```python
df = df[df['Price'] != 'twfqy']
```

```python
df['Price'] = pd.to_numeric(df["Price"])
```

```python
df.head()
```

```python
df.info()
```

```python
df['X'] = df['Address'].apply(lambda x : x.split()[-2:])
```

```python
df['X'] = df['X'].apply(lambda x: ','.join(map(str, x)))
df.head()
```

```python
df['X'] = df['X'].str.replace(',','  ')
df
```

```python
df['X'] = df['X'].str.replace('(','')
df['X'] = df['X'].str.replace(')','')
df
```

```python
df['X'] = df['X'].str.replace('تهران','')
df
```

```python
df.drop('Address',inplace=True,axis=1)
df
```

```python
df.rename(columns={'X':'Address'},inplace=True)
df
```

```python
mapping = {'True':1, 'False':0}
df['Warehouse'] = df['Warehouse'].map(mapping)
df['Elevator'] = df['Elevator'].map(mapping)
df['Parking'] = df['Parking'].map(mapping)
```

```python
df['Area'] = df['Area'].astype(int)
df['Construction'] = df['Construction'].astype(int)
df['Room'] = df['Room'].astype(int)
```
